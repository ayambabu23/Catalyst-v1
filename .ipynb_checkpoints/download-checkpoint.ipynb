```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
import tensorflow as tf
from yolov3.yolov4 import Create_Yolo
from yolov3.utils import load_yolo_weights, detect_image
from yolov3.configs import *
#%cd ../../..
#%cd TensorFlow-2.x-YOLOv3
#%cd IMAGES/footage/
vidcap = cv2.VideoCapture('./IMAGES/footage/Kuldeep Bowls Babar.mov')
success,image = vidcap.read()
count = 0
yolo = Create_Yolo(input_size=YOLO_INPUT_SIZE, CLASSES=TRAIN_CLASSES)
yolo.load_weights("./checkpoints/yolov3_custom")

while success:
    %cd IMAGES/footage/frames
    image_name = "frame%d.jpg" % count
    print("FRAME %d:" % count)
    cv2.imwrite(image_name, image)     # save frame as JPEG file
    %cd ../../..
    image = detect_image(yolo, "./IMAGES/footage/frames/"+image_name, "", input_size=YOLO_INPUT_SIZE, show=False, CLASSES=TRAIN_CLASSES, rectangle_colors=(255,0,0))
    #%cd ../../..
    success,image = vidcap.read()
    #print('Read a new frame: ', success)
    count += 1
```
